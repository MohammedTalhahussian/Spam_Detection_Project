# Spam Detection using Machine Learning

This notebook builds and evaluates supervised classification models for spam detection.

```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)
```

## Load Dataset and Preprocess Data

```

df = pd.read_csv("spam.xls.csv", encoding='latin-1')

df = df[['v1', 'v2']]

df.columns = ['label', 'message']

df['label'] = df['label'].map({
    'ham': 0,
    'spam': 1
})

X = df['message']
y = df['label']

vectorizer = TfidfVectorizer(
    stop_words='english',
    max_features=3000
)

X_tfidf = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print(df.head())
print("\nDataset Shape:", df.shape)
```

## Logistic Regression Model

```

log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train, y_train)

y_pred_log = log_model.predict(X_test)

log_accuracy = accuracy_score(y_test, y_pred_log)
log_precision = precision_score(y_test, y_pred_log)
log_recall = recall_score(y_test, y_pred_log)
log_f1 = f1_score(y_test, y_pred_log)
log_roc_auc = roc_auc_score(y_test, y_pred_log)

print("Accuracy:", log_accuracy)
print("Precision:", log_precision)
print("Recall:", log_recall)
print("F1 Score:", log_f1)
print("ROC-AUC:", log_roc_auc)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_log))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_log))
```

## Random Forest Model

```

rf_model = RandomForestClassifier()

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_precision = precision_score(y_test, y_pred_rf)
rf_recall = recall_score(y_test, y_pred_rf)
rf_f1 = f1_score(y_test, y_pred_rf)
rf_roc_auc = roc_auc_score(y_test, y_pred_rf)

print("Accuracy:", rf_accuracy)
print("Precision:", rf_precision)
print("Recall:", rf_recall)
print("F1 Score:", rf_f1)
print("ROC-AUC:", rf_roc_auc)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))
```

## Cross Validation

```

log_cv = cross_val_score(
    log_model,
    X_tfidf,
    y,
    cv=5,
    scoring='accuracy'
)

rf_cv = cross_val_score(
    rf_model,
    X_tfidf,
    y,
    cv=5,
    scoring='accuracy'
)

print("Logistic Regression CV Mean:", log_cv.mean())
print("Random Forest CV Mean:", rf_cv.mean())
```

## Metrics Comparison Table

```

metrics_df = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest'],
    'Accuracy': [log_accuracy, rf_accuracy],
    'Precision': [log_precision, rf_precision],
    'Recall': [log_recall, rf_recall],
    'F1 Score': [log_f1, rf_f1],
    'ROC-AUC': [log_roc_auc, rf_roc_auc]
})

print(metrics_df)
```

## Accuracy Plot

```

models = ['Logistic Regression', 'Random Forest']
accuracy_scores = [log_accuracy, rf_accuracy]

plt.figure(figsize=(8,5))

plt.bar(models, accuracy_scores)

plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Accuracy Comparison")

plt.show()
```

## Precision Plot

```

precision_scores = [log_precision, rf_precision]

plt.figure(figsize=(8,5))

plt.bar(models, precision_scores)

plt.xlabel("Models")
plt.ylabel("Precision")
plt.title("Precision Comparison")

plt.show()
```

## Recall Plot

```

recall_scores = [log_recall, rf_recall]

plt.figure(figsize=(8,5))

plt.bar(models, recall_scores)

plt.xlabel("Models")
plt.ylabel("Recall")
plt.title("Recall Comparison")

plt.show()
```

## F1 Score Plot

```

f1_scores = [log_f1, rf_f1]

plt.figure(figsize=(8,5))

plt.bar(models, f1_scores)

plt.xlabel("Models")
plt.ylabel("F1 Score")
plt.title("F1 Score Comparison")

plt.show()
```
